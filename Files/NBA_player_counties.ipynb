```python
import pandas as pd
import requests
import json
import time
```

```python
from config import gkey
```

```python
geo_url = 'https://maps.googleapis.com/maps/api/geocode/json?'
geo_params = {
    'address':'Austin, Texas',
    'key':gkey
}
```

```python
url = "https://geo.fcc.gov/api/census/area?"

params = {
    'lat':30.266666,
    'lon':-97.733330,
    'format':'json'
}
```

```python
csv_path = "../Resources/clean_data/US_players_coordinates.csv"
df = pd.read_csv(csv_path)
df.head()
```

```python
birth_county = []
```

```python
for index, row in df.iterrows():
    params['lat'] = row['Latitude (generated)']
    params['lon'] = row['Longitude (generated)']
    
    try:
        response = requests.get(url, params).json()
        birth_county.append(response['results'][0]['county_name'])
        time.sleep(0.05)
    except:
        geo_params['address'] = f"{row['Birth City']}, {row['Birth State']}"
        geo_response = requests.get(geo_url, geo_params).json()
        df.at[index, 'Latitude (generated)'] = geo_response["results"][0]["geometry"]["location"]["lat"]
        df.at[index, 'Longitude (generated)'] = geo_response["results"][0]["geometry"]["location"]["lng"]
        
        params['lat'] = geo_response["results"][0]["geometry"]["location"]["lat"]
        params['lon'] = geo_response["results"][0]["geometry"]["location"]["lng"]
        try:
            response = requests.get(url, params).json()
            birth_county.append(response['results'][0]['county_name'])
            time.sleep(0.05)
        except Exception as e:
            print(index)
            print(e)
        print(index)
```

```python
df.to_csv("../Resources/clean_data/US_players_coordinates.csv", header=True)
```

```python
df["Birth County"] = birth_county
```
